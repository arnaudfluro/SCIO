```python
import requests as req         # module de requête http
url="http://www.assemblee-nationale.fr/dyn/opendata/PRJLANR5L15B2488.html"
page = req.request('get', url) # on récupère une page 
html = page.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html)
```

```python
#On récupérer les balises ouil y a le texte avec les articles
selector = "assnat9ArticleNum" #titre des articles
nodes_title_article    = soup.find_all(class_=selector)
len(nodes_title_article)
```

```python
import re
titles_article=[]
for k in range(len(nodes_title_article)):
    titles_article.append(re.findall("\n(.*?)\n",nodes_title_article[k].text)[0])
```

```python
#Récupération des paragraphes lié à l'article class=assnatLoiTexte (paragraphes situé entre deux articles)
import pandas as pd
intitulte_article_df=pd.DataFrame()
intitulte_article_df['Article']=titles_article
```

```python
# por _ in liste article tant que selector(p)
selector = "p" #titre des articles
nodes_p= soup.find_all(selector)
len(nodes_p)
```

```python
nodes_p
```

```python
#L'index de chaque article
index_article=[] #retrouver les index de nos articles afin de capturer l'intitulé de chaque article
selector = "assnat9ArticleNum"
for (index,texte) in  enumerate(nodes_p):
    soup = BeautifulSoup(str(texte))
    test=soup.find(class_=selector)
    if test!=None:
        print(test.text)
        index_article.append(index)
```

```python
len(index_article) #on a bien l'index de tout les articles
```

```python
nodes_p[index_article[0]+1].text
```

```python
nb_article=len(index_article)
intitule_articles=[]
selector="assnatLoiTexte" #que l'intitulé des articles et non les titres,...
for index in range(nb_article-1):
    intitulte_article_k=""
    for k in range(index_article[index]+1,index_article[index+1]): #le 1er +1 nous permet de prendre ce qu'il y a à l'intérieur 
        soup=BeautifulSoup(str(nodes_p[k]))
        intitule_text=soup.find(class_=selector)
        if intitule_text!=None: #les noeuds qui n'ont pas la class=selector
            intitulte_article_k+=intitule_text.text
    intitule_articles.append(intitulte_article_k)

#Récupération dernier article
for k in range(nb_article,len(nodes_p)):
    soup=BeautifulSoup(str(nodes_p[k]))
    intitule_text=soup.find(class_=selector)
    if intitule_text!=None: 
        intitulte_article_k+=intitule_text.text
intitule_articles.append(intitulte_article_k)
```

```python
def clean_texte(texte):
    texte_clean=texte.strip().replace('\n', '')
    texte_clean=texte_clean.strip().replace('\xa0', ' ')
    texte_clean=texte_clean.strip().replace('\t', '')
    return(texte_clean)
```

```python
#Nettoyage des \xa0 \n\t
def cleaning_article(texte_article):
    for (index,texte) in enumerate(texte_article):
        if texte!=None:              
            texte_clean=clean_texte(texte)
            texte_article[index]=texte_clean
```

```python
#ne fonctionne pas sans cleaning
#Recupere la liste des numéros d'article concernés
def xx_a_yy(string):
    liste_article=[]
    string=clean_texte(string)
    match=re.findall('\d+ à \d+',string)
    #print(match[0])
    if(match!=None):
        match_only_int=[int(i) for i in re.findall('\d+',match[0])]
    return [k for k in range(match_only_int[0],match_only_int[1]+1) ] #dernier chifre inclut
```

```python
list_article=[]
num_article_dans_paragraphe=[]
for node_p in nodes_p:
    for child in node_p.children:
        try:
            if child.name=='span':
                #On va macher le numéro de l'article en gras à l'article dans intitule article_df
                #Si on trouve on rajoute le/les paragraphes concerné par les articles
                for span in child: #span = contenu en gras 
                    #quel article on parle dans le pagraphe (toujours en gras)
                    num_article_dans_paragraphe=[int(re.match('\d+',span)[0])]
                    for i in num_article_dans_paragraphe:
                        intitulte_article_df['Simplifie'][i-1]=node_p.text
                    
                    #On traite le cas si 77 à 81 par exemple
                    num_article_xx_to_yy=xx_a_yy(node_p.text)
                    for i in num_article_xx_to_yy:
                        intitulte_article_df['Simplifie'][i-1]=node_p.text
                
                
        except Exception as e:
            pass
```
